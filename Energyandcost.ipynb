Energy-Aware Scheduling

```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

Define carbon intensity (grams of CO2 per kWh) for 3 regions



```python
regions = {
"US-West": {"carbon_intensity": 400, "cost_per_kwh": 0.12},
"EU-Central": {"carbon_intensity": 250, "cost_per_kwh": 0.15},
"Asia-East": {"carbon_intensity": 600, "cost_per_kwh": 0.10}
}
```

Simulate 100 jobs with random energy demand (kWh)

```python
num_jobs = 100
job_energy = [random.uniform(0.5, 3.0) for _ in range(num_jobs)]
```

Greedy scheduler (assigns based on lowest cost)

```python
greedy_schedule = [min(regions, key=lambda r: regions[r]["cost_per_kwh"]) for _ in job_energy]
greedy_emissions = [job_energy[i] * regions[greedy_schedule[i]]["carbon_intensity"] for i in range(num_jobs)]
```

Energy-aware scheduler (assigns based on lowest carbon intensity)

```python
green_schedule = [min(regions, key=lambda r: regions[r]["carbon_intensity"]) for _ in job_energy]
green_emissions = [job_energy[i] * regions[green_schedule[i]]["carbon_intensity"] for i in range(num_jobs)]
```

Compute totals

```python
total_greedy = sum(greedy_emissions)
total_green = sum(green_emissions)
```

Bar chart comparison

```python
labels = ['Greedy Scheduler', 'Energy-Aware Scheduler']
values = [total_greedy, total_green]
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['red', 'green'])
plt.ylabel("Total CO₂ Emissions (grams)")
plt.title("CO₂ Emissions Comparison: Greedy vs. Energy-Aware Scheduler")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
print("Total CO2 Emissions (Greedy Scheduler):", total_greedy, "grams")
print("Total CO2 Emissions (Energy-Aware Scheduler):", total_green, "grams")
```
